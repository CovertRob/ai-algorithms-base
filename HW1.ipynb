## Background on Algorithm Analysis (50 points total)

- Constant Time (O(1)): Regardless of the input size, the algorithm takes a constant amount of time to complete.
    - Accessing an element in an array by index.
    - Inserting or deleting an element at the beginning of a linked list.


- Linear Time (O(n)): The runtime grows linearly with the size of the input.
    - Iterating through a list or array once.
    - Searching for an element in an unsorted list by traversing it.


- Logarithmic Time (O(log n)): As the input size grows, the runtime increases logarithmically.
    - Binary search in a sorted array.
    - Operations in a balanced binary search tree (BST).
    

- Quadratic Time (O(n^2)): The runtime grows quadratically with the input size.
    - Nested loops where each loop iterates through the input.
    - Some sorting algorithms like Bubble Sort or Selection Sort.


- Exponential Time (O(2^n)): The runtime doubles with each addition to the input size.
    - Algorithms involving recursive solutions that make repeated calls.
    - Solving the Traveling Salesman Problem using brute force.
    

- Factorial Time (O(n!)): The runtime grows factorially with the input size.
    - Permutation problems that explore all possible combinations, like the brute force solution for the Traveling Salesman Problem with no optimizations.

## Bubble Sort (Example)

Use the below analysis and explanation as a template for answering the following questions on algorithm analysis. You are to analyze each algorithm line by line, calculate the runtime complexity, and provide an explanation in markdown.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):  # O(n)
        for j in range(0, n-i-1):  # O(n)
            if arr[j] > arr[j+1]:  # O(1)
                arr[j], arr[j+1] = arr[j+1], arr[j]  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n^2)
# Big Omega notation: Ω(n)
# Big Theta notation: Θ(n^2)
arr = [1, 2, 5, 3, 43]
bubble_sort(arr)
print(arr)
```

Explanation:

- The outer loop iterates 'n' times, where 'n' is the length of the array.
- The inner loop also iterates 'n' times in the worst case.
- The comparisons and swapping operations inside the inner loop are constant time ('O(1)').

This results in a total runtime complexity of O(n^2) in the worst case, Ω(n) in the best case (already sorted), and Θ(n^2) in the average and worst cases, as both loops iterate over the input array.

## 1A Merge Sort (10 points)

```python
def merge_sort(arr):
    if len(arr) > 1:  # O(1)
        mid = len(arr) // 2  # O(1)
        left_half = arr[:mid]  # O(n)
        right_half = arr[mid:]  # O(n)

        merge_sort(left_half)  # T(n/2)
        merge_sort(right_half)  # T(n/2)

        i = j = k = 0  # O(1)

        while i < len(left_half) and j < len(right_half):  # O(n)
            if left_half[i] < right_half[j]:  # O(1)
                arr[k] = left_half[i]  # O(1)
                i += 1  # O(1)
            else:
                arr[k] = right_half[j]  # O(1)
                j += 1  # O(1)
            k += 1  # O(1)

        while i < len(left_half):  # O(n)
            arr[k] = left_half[i]  # O(1)
            i += 1  # O(1)
            k += 1  # O(1)

        while j < len(right_half):  # O(n)
            arr[k] = right_half[j]  # O(1)
            j += 1  # O(1)
            k += 1  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n log n)
# Big Omega notation: Ω(n log n)
# Big Theta notation: Θ(n log n)
```

Explanation:

- The array is split in half each level (two recursive calls on n/2), giving log n levels.
- Each level performs linear work to merge the halves (the while loops total O(n) per level).
- Therefore the overall runtime is O(n log n) in the worst, best, and average cases.


## 1B Binary Search (10 points)

```python
def binary_search(arr, target):
    low = 0  # O(1)
    high = len(arr) - 1  # O(1)

    while low <= high:  # O(log n)
        mid = (low + high) // 2  # O(1)
        mid_val = arr[mid]  # O(1)

        if mid_val == target:  # O(1)
            return mid  # O(1)
        elif mid_val < target:  # O(1)
            low = mid + 1  # O(1)
        else:
            high = mid - 1  # O(1)

    return -1  # O(1)

# Total Runtime Complexity:
# Big O notation: O(log n)
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(log n)
```

Explanation:

- Each loop iteration cuts the search space in half, so the number of iterations is O(log n).
- Work inside the loop is constant time.
- Best case is Ω(1) if the target is found at the first mid; otherwise the runtime is Θ(log n).


## 1C Linear Search (10 points)

```python
def linear_search(arr, target):
    for i in range(len(arr)):  # O(n)
        if arr[i] == target:  # O(1)
            return i  # O(1)
    return -1  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n)
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(n)
```

Explanation:

- The loop may check every element in the array in the worst case, giving O(n).
- Each comparison is constant time.
- Best case is Ω(1) if the target is at the first position; average and worst are Θ(n).


## 1D Selection Sort (10 points)

```python
def selection_sort(arr):
    n = len(arr)  # O(1)
    for i in range(n - 1):  # O(n)
        min_index = i  # O(1)
        for j in range(i + 1, n):  # O(n)
            if arr[j] < arr[min_index]:  # O(1)
                min_index = j  # O(1)
        arr[i], arr[min_index] = arr[min_index], arr[i]  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n^2)
# Big Omega notation: Ω(n^2)
# Big Theta notation: Θ(n^2)
```

Explanation:

- The outer loop runs n times and the inner loop scans the remaining elements each time.
- The nested loops produce n + (n-1) + ... + 1 comparisons, which is O(n^2).
- Selection sort does the same comparisons regardless of input order, so best and average are also Θ(n^2).


## 1E Heap Sort (10 points)

```python
def heapify(arr, n, i):
    largest = i  # O(1)
    left = 2 * i + 1  # O(1)
    right = 2 * i + 2  # O(1)

    if left < n and arr[left] > arr[largest]:  # O(1)
        largest = left  # O(1)

    if right < n and arr[right] > arr[largest]:  # O(1)
        largest = right  # O(1)

    if largest != i:  # O(1)
        arr[i], arr[largest] = arr[largest], arr[i]  # O(1)
        heapify(arr, n, largest)  # O(log n)


def heap_sort(arr):
    n = len(arr)  # O(1)

    for i in range(n // 2 - 1, -1, -1):  # O(n)
        heapify(arr, n, i)  # O(log n)

    for i in range(n - 1, 0, -1):  # O(n)
        arr[i], arr[0] = arr[0], arr[i]  # O(1)
        heapify(arr, i, 0)  # O(log n)

# Total Runtime Complexity:
# Big O notation: O(n log n)
# Big Omega notation: Ω(n log n)
# Big Theta notation: Θ(n log n)
```

Explanation:

- Building the heap is linear to near-linear work overall; each heapify call is O(log n).
- The extraction loop runs n times, and each extraction costs O(log n) to re-heapify.
- This yields O(n log n) in the worst, best, and average cases.

```python
# Prime Number Checking (Trial Division)
# Short and sweet so easier to just do the code directly

def is_prime(n):
    if n <= 1:  # O(1)
        return False  # O(1)
    if n == 2:  # O(1)
        return True  # O(1)
    if n % 2 == 0:  # O(1)
        return False  # O(1)

    i = 3  # O(1)
    while i * i <= n:  # O(sqrt(n)) ... easy speed up by implicity calculating sqrt(n) and avoiding a library call
        if n % i == 0:  # O(1)
            return False  # O(1)
        i += 2  # O(1)

    return True  # O(1)

# Total Runtime Complexity:
# Big O notation: O(sqrt(n))
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(sqrt(n))
```

Explanation:

- I start by handling the small and obvious cases (n <= 1, n == 2, even numbers). Each of those checks is constant time.
- The trial division loop only tests odd divisors up to sqrt(n). If n has a factor, it must appear at or before sqrt(n), so we never need to go past that.
- The loop advances by 2 each time, so in the worst case I do about sqrt(n)/2 iterations, which is still Θ(sqrt(n)).
- Best case is Ω(1) when n is small or when an early check (like evenness) ends the function right away.

**Discussion**

For space, an adjacency matrix costs O(V^2) no matter how many edges you actually have, so it’s great for dense graphs but wasteful for sparse ones. An adjacency list costs O(V + E), which is usually much smaller when the graph is sparse, like this one. Operationally, a matrix gives O(1) edge lookups (just check a cell), while a list gives O(deg(v)) lookups but is faster for iterating over real neighbors because you only traverse existing edges.

That choice matters for algorithms: BFS/DFS typically run faster in practice with adjacency lists on sparse graphs because they only touch real edges. On the other hand, some optimization routines or dense-graph checks can benefit from a matrix’s constant‑time edge existence tests. So for this graph, the list is more space‑efficient and a natural fit for traversals, while the matrix is handy when I need quick edge checks or a clean, tabular representation.

```python
# Q3B: BFS & DFS

from collections import deque
import matplotlib.pyplot as plt
import networkx as nx

vertices = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]

edges = [
    # Unweighted + undirected
    ("A", "B"),
    ("A", "C"),
    ("B", "C"),
    ("B", "D"),
    ("B", "E"),
    ("C", "E"),
    ("D", "E"),
    ("D", "F"),
    ("E", "F"),
    ("E", "G"),
    ("G", "H"),
    ("G", "J"),
    ("J", "I"),
    ("H", "I"),
    ("F", "H"),
    ("E", "H"),
]

# Build adjacency list (alphabetical)
# sort neighbors so ties break alphabetically
adj = {v: [] for v in vertices}
for u, v in edges:
    adj[u].append(v)
    adj[v].append(u)
for v in vertices:
    adj[v].sort()

# BFS from A
# BFS: explore layer by layer to get shortest-hop levels.
start = "A"
visited = set([start])
queue = deque([start])
# For keeping track of the order of nodes we visit
order_bfs = []
# For keeping track of a nodes parents
parent_bfs = {start: None}
# For keeping track of a nodes levels
level = {start: 0}

while queue:
    u = queue.popleft()
    order_bfs.append(u) # For tracking
    for v in adj[u]:
        if v not in visited:
            visited.add(v)
            parent_bfs[v] = u # For tracking
            level[v] = level[u] + 1 # For tracking
            queue.append(v)

bfs_tree_edges = [(p, v) for v, p in parent_bfs.items() if p is not None]

print("BFS visitation order:", order_bfs)
print("BFS tree edges:", bfs_tree_edges)
print("BFS levels:", level)

# DFS from A (preorder)
# Pre-order records a node when you first visit it (on entry) aka Parent before children
# Post-order records a node after you've finished visiting all its descendants (on exit) aka Children before Parent
# DFS: go deep first, preorder is the visit order as I dive.
visited = set()
order_dfs = []
dfs_tree_edges = []

# Recursive DFS function
def dfs(u):
    visited.add(u)
    order_dfs.append(u)
    for v in adj[u]:
        if v not in visited:
            dfs_tree_edges.append((u, v))
            dfs(v)

dfs(start)

print("\nDFS preorder:", order_dfs)
print("DFS tree edges:", dfs_tree_edges)

# Visualize BFS and DFS trees
# I’m drawing the original graph in light gray and then overlaying each tree.
G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(edges)

# Layout
pos = {
    "A": (0.0, 3.0),
    "B": (-1.3, 2.0),
    "C": (1.3, 2.0),
    "D": (-1.3, 1.0),
    "E": (1.3, 1.0),
    "F": (-1.3, 0.0),
    "H": (2.4, 0.2),
    "G": (3.2, 1.0),
    "J": (4.5, 1.0),
    "I": (3.4, -0.5),
}

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("BFS Tree (start A)")
# draw all edges light gray
nx.draw(G, pos=pos, with_labels=True, node_size=700, edge_color="#d0d0d0", width=1)
# overlay BFS tree edges
nx.draw_networkx_edges(G, pos=pos, edgelist=bfs_tree_edges, edge_color="#1f77b4", width=2.5)

plt.subplot(1, 2, 2)
plt.title("DFS Tree (start A)")
# draw all edges light gray
nx.draw(G, pos=pos, with_labels=True, node_size=700, edge_color="#d0d0d0", width=1)
# overlay DFS tree edges
nx.draw_networkx_edges(G, pos=pos, edgelist=dfs_tree_edges, edge_color="#d62728", width=2.5)

plt.tight_layout()
plt.show()
```

**Analysis**

BFS guarantees shortest‑hop paths in unweighted graphs because it explores the graph level by level, so the first time you reach a node, you’ve used the fewest edges possible. DFS doesn’t give that guarantee because it dives deep along one path before backtracking, so it can reach a node via a longer route even if a shorter route exists. A good AI use case for BFS is finding the minimum number of moves in a state space (like shortest‑step pathfinding on an unweighted grid). A good AI use case for DFS is exploring a large decision space where you want to find any solution quickly or do systematic backtracking (like constraint satisfaction search).

```python
# Q3C: Minimum Spanning Tree using Prim's Algorithm

import heapq

vertices = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]

edges = [
    ("A", "B", 2),
    ("A", "C", 5),
    ("B", "C", 2),
    ("B", "D", 1),
    ("B", "E", 2),
    ("C", "E", 3),
    ("D", "E", 1),
    ("D", "F", 4),
    ("E", "F", 2),
    ("E", "G", 6),
    ("G", "H", 1),
    ("G", "J", 5),
    ("J", "I", 2),
    ("H", "I", 4),
    ("F", "H", 3),
    ("E", "H", 1),
]

# Build adjacency list with weights
adj = {v: [] for v in vertices}
for u, v, w in edges:
    adj[u].append((v, w))
    adj[v].append((u, w))

start = "A"
visited = set([start])
frontier = []
frontier_steps = []

# seed frontier with edges from start
for v, w in adj[start]:
    heapq.heappush(frontier, (w, start, v))

mst_edges = []
total_weight = 0

step = 1
while frontier and len(visited) < len(vertices):
    w, u, v = heapq.heappop(frontier)
    if v in visited:
        continue
    visited.add(v)
    mst_edges.append((u, v, w))
    total_weight += w

    # record the chosen frontier edge here
    frontier_steps.append((step, u, v, w))
    step += 1

    for nxt, wt in adj[v]:
        if nxt not in visited:
            heapq.heappush(frontier, (wt, v, nxt))

print("Prim start:", start)
print("Frontier edges chosen by step:")
for step_num, u, v, w in frontier_steps:
    print(f"  Step {step_num}: ({u}, {v}, {w})")

print("MST edges:", mst_edges)
print("Total weight:", total_weight)
```


MSTs aren’t always unique. When multiple edges share the same weight, different (but equally valid) choices can lead to different MSTs with the same total cost.

**Why MSTs in AI?**

A minimum spanning tree connects all vertices with the smallest total edge cost, which is different from a shortest‑path tree that optimizes paths from a single source. In AI, MSTs show up in clustering , network design, and building sparse structures that still preserve connectivity. The key idea is global efficiency, not shortest routes from one start node.
