##Binary Search - iterative

```python
def binary_search(arr, target):
    """Binary search in sorted array."""
    left, right = 0, len(arr) - 1

    while left <= right:             # Line 1: while loop
        mid = (left + right) // 2    # Line 2: calculate midpoint
        if arr[mid] == target:       # Line 3: comparison (basic operation)
            return mid
        elif arr[mid] < target:      # Line 4: comparison
            left = mid + 1           # Line 5: update left
        else:
            right = mid - 1          # Line 6: update right

    return -1
```

### Analysis

**Basic Operation:** Comparisons at Lines 3 and 4
<br>
**Input Size**: n = len(arr)<br>
**Loop Analysis**:<br>
*   Each iteration eliminates half of the remaining elements
*   Search space: n → n/2 → n/4 → n/8 → ... → 1
*   Number of iterations until search space ≤ 1: ⌊log₂(n)⌋ + 1

**Worst Case**: The target isn't found or it's at a position requiring maximum divisions
$$
T(n) = \lfloor \log_2(n) \rfloor + 1 \in O(\log n)
$$


## Recursive example - Simple Linear Recursion - Factorials


```python
def factorial(n):
    """Calculate n! using recursion."""
    if n <= 1:                       # Line 1: base case condition
        return 1                     # Line 2: base case return
    else:
        return n * factorial(n - 1)  # Line 3: recursive call
```

###Analysis
**Basic Operation**: The multiplication $n * factorial(n-1)$ <br>
**Setting up the Recurrence**:<br>
*   Let T(n) = number of multiplications to compute factorial(n)
*   Base case: T(0) = T(1) = 0 (no multiplications needed)
*   Recursive case: T(n) = 1 + T(n - 1) for n > 1

**Solving the Recurrence**:<br>
$$
\begin{align}
T(n) &= 1 + T(n - 1) \\
T(n-1) &= 1 + T(n - 2) \\
T(n-2) &= 1 + T(n - 3) \\
&\vdots \\
T(2) &= 1 + T(1) \\
T(1) &= 0
\end{align}
$$
<br>
**Substitution Method**:<br>
$$
\begin{align}
T(n) &= 1 + T(n - 1) \\
     &= 1 + (1 + T(n - 2)) \\
     &= 1 + 1 + (1 + T(n - 3)) \\
     &= 1 + 1 + 1 + ... + T(1) \\
     &= (n-1) + 0 \\
     &= n-1
\end{align}
$$
<br>
**Final Complexity**:
$$
T(n) = n-1 \in O(n)
$$
<br>
**Call Stack Analysis**:<br>

*   Maximum recursion depth: n
*   Space complexity: **O(n)** due to call stack
