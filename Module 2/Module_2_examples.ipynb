# Dijkstra's Algorithm 
## Smart City Traffic Routing
### Problem: 
An emergency vehicle needs optimal routing through a city network with real-time traffic delays. Find the shortest path from Hospital A to all other intersections.<br>

#### Network Setup:<br>
![Module 2 example 1](<Module 2 example 1 white bg.jpg>)

#### Graph Properties:<br>
* Vertices: V = {A, B, C, D, E}  representing intersections <br>
* Edges: Roads with travel time weights (all non-negative) <br>
* Source: A (Hospital)<br>
* Goal: Find shortest paths to all vertices<br>
<br>


#### Math:
Dijkstra's algorithm is based on the optimal substructure property of shortest paths. If the shortest path from source _s_ to vertex _v_ goes through vertex _u_, then the path from _s_ to _u_ must also be shortest. 
##### Key: 
Once a vertex is extracted from the priority queue, its shortest distance from the source is finalized and will never change.

##### Why This Works: 
Since all edge weights are non-negative, any future path to a processed vertex will have distance ≥ current shortest distance.

### Algorithm:
Initialization:
* dist[A] = 0 (source node - distance to itself is 0)<br>
* dist[v] = ∞ for all v ≠ A (unknown distances start as infinite)<br>
* Priority Queue Q = {(A, 0)} (stores vertices by their current best distance)
* parent[v] = null for all v (for path reconstruction)
Relaxation Formula: For each edge (u,v) with weight w<br>
```
if dist[u] + w < dist[v]:<br>
    dist[v] = dist[u] + w    		# Update shorter distance found
    parent[v] = u            		# Track path for reconstruction
    update v in priority queue  	# May decrease key in min-heap
```      

In Dijkstra's algorithm, relaxation is the process of checking whether we can improve (reduce) the shortest known distance to a vertex by going through another vertex. Think of "relaxation" like loosening a tense rubber band or spring. Initially, all distances are "tense" (set to infinity, meaning we don't know how to reach them). When we find a shorter path, we "relax" that distance by making it smaller.

#### Why It's Called "Relaxation":
The term comes from optimization theory. We start with "tight" constraints (distances set to ∞), and gradually "relax" them by finding shorter paths. Each relaxation makes our estimate less restrictive and more accurate.
#### Why Relaxation Works: 
If we've found the shortest path to u, and there's an edge u → v, then dist[u] + weight(u, v) represents a candidate shortest path to v. We only update if this candidate is better than our current best.


#### Implementation:
##### Initial State:
```
dist = {A:0, B:∞, C:∞, D:∞, E:∞}
parent = {A:null, B:null, C:null, D:null, E:null}
Q = [(0, A)]          # priority queue with (distance, vertex) pairs
```

1. Extract minimum: (0, A) <br> 
*Why A? It has the smallest distance (0) in the priority queue*
* Mark A as processed since its shortest distance is now final
* Examine A's neighbors: B and C
*  Relax A → B: dist[B] = min(∞, 0 + 2) = 2
    * Reasoning: Path A → B has cost 2, which is better than ∞
    * parent[B] = A, add (2, B) to queue
* Relax A→C: dist[C] = min(∞, 0 + 4) = 4
    * Reasoning: Path A → C has cost 4, which is better than ∞
    * parent[C] = A, add (4, C) to queue
* Updated state: dist = {A:0, B:2, C:4, D:∞, E:∞}
* Q = [(2, B), (4, C)]
2. Extract minimum: (2, B) <br> 
*Why B? It has distance 2, smaller than C's distance 4*
* Mark B as processed since distance 2 to B is now final and optimal
* Examine B's neighbors: D only
* Relax B → D: dist[D] = min(∞, 2 + 3) = 5
    * Reasoning: Path A → B → D has total cost 5, better than ∞
    * parent[D] = B, add (5, D) to queue
* Updated state: dist = {A:0, B:2, C:4, D:5, E:∞}
* Q = [(4, C), (5, D)]
3. Extract minimum: (4, C) <br>	
*Why C? It has distance 4, smaller than D's distance 5*
* Mark C as processed since distance 4 to C is now final
* Examine C's neighbors: E only
* Relax C → E: dist[E] = min(∞, 4 + 1) = 5
    * Reasoning: Path A → C → E has total cost 5, better than ∞
    * parent[E] = C, add (5, E) to queue
* Updated state: dist = {A:0, B:2, C:4, D:5, E:5}
* Q = [(5, D), (5, E)] <br>		
*Note: Both D and E have the same distance!*
4. Extract minimum: (5, D) 
*Note: D and E both have distance 5 so we can extract either. Assume min-heap returns D first*
* Mark D as processed since distance 5 to D is now final
* Examine D's neighbors: E only
* Relax D → E: dist[E] = min(5, 5 + 5) = 5 (no improvement)
    * Reasoning: Path A → B → D → E costs 5 + 5 = 10, worse than current path A → C → E costing 5
    * Insight: This shows why we don't always update - sometimes longer paths cost more
* Updated state: No changes
* Q = [(5, E)]
5. Extract minimum: (5, E)
* Mark E as processed since distance 5 to E is now final
* E has no outgoing edges in this example
* Q = [] (empty - algorithm terminates)

### Algorithm Correctness Proof:
* When we process a vertex *u*, dist[*u*] is the shortest distance from source to *u*
* This is because all remaining vertices in the queue have distance ≥ dist[*u*], and all edge weights are non-negative
* Therefore, no future relaxation can improve dist[*u*]
#### Final Results:
* Shortest Distances: A(0), B(2), C(4), D(5), E(5)
* Optimal Path to E: A → C → E (total cost: 5)
* Path Reconstruction: Follow parent pointers: E ← C ← A gives path A → C → E
* Alternative Path: A → B → D costs 5, and A → B → D → E costs 10 (suboptimal)

### Time Complexity Analysis:
* With binary heap: O((*V + E*) *log V*)
    * V extractions from heap: O(*V log V*)
    * E edge relaxations with heap updates: O(*E log V*)
* Space Complexity: O(*V*) for distance array, parent array, and priority queue
#### Why Dijkstra's Fails with Negative Weights: 
The algorithm assumes that once a vertex is processed (removed from the queue), its shortest distance is final. With negative weights, a path that's discovered later might be shorter, violating this assumption


# Bellman-Ford with Negative Weights
## Cryptocurrency Arbitrage Detection
### Problem: 
In cryptocurrency arbitrage, we're looking for trading cycles where we end up with more money than we started with. 
#### Real-World Context: 
Cryptocurrency arbitrage involves exploiting price differences across exchanges. When converted to logarithmic space, multiplicative gains become additive, and profitable cycles appear as negative-weight cycles
#### Exchange Network:
![Module 2 example 2](<Module 2 example 2 white bg.jpg>)

#### Math: 
*Why do we need a logarithmic transformation?*
Graph algorithms like Bellman-Ford are designed to add edge weights along paths, not multiply them. The algorithms look for shortest paths by adding up edge costs.
But we need to find cycles where multiplying exchange rates gives us a value greater than 1.
##### The Transformation in Practice:
* Original problem: Find cycles where (rate₁ × rate₂ × rate₃...) > 1
* After taking logs: Find cycles where (log(rate₁) + log(rate₂) + log(rate₃)...) > 0
* After negating: Find cycles where (-log(rate₁) + -log(rate₂) + -log(rate₃)...) < 0
* This is exactly what Bellman-Ford detects: negative-weight cycles
##### Edge Weight Interpretation:
* Positive weight: -log(rate) > 0 means rate < 1 (losing money in this trade)
* Negative weight: -log(rate) < 0 means rate > 1 (gaining money in this trade)
* Zero weight: Rate = 1 (break even trade)
##### Example: 
*ETH → SOL* edge has weight -1<br>
* This means: -log(exchange_rate) = -1
* Therefore: log(exchange_rate) = 1
* Therefore: exchange_rate = e¹ ≈ 2.718
* Interpretation: 1 *ETH* converts to ~2.718 *SOL* (profitable)
##### Algorithm Theory:
Unlike Dijkstra's, Bellman-Ford can handle negative weights because it doesn't use a greedy approach. Instead, it systematically relaxes all edges multiple times.
##### Insight: 
In a graph with n vertices, the shortest path between any two vertices contains at most *n - 1* edges (assuming no negative cycles). Therefore, after *n - 1* iterations of relaxing all edges, we've found all shortest paths. If we can still improve distances in iteration *n*, there must be a negative cycle.


### Implementation:
#### Initialization:
```
dist[*BTC*] = 0    		# Source cryptocurrency (our starting point)
dist[*v*] = ∞      		# All other cryptocurrencies initially unreachable
parent[*v*] = null 		# For cycle reconstruction
```

**Main Algorithm (|*V*| - 1 = 4 iterations for 5 vertices):**
##### Iteration 1:
First Round of Relaxations: <br>
*Goal: Find all paths of length 1 from source*
Process ALL edges in any order:
* *BTC → ETH*(2): 
    * dist[*ETH*] = min(∞, 0 + 2) = 2
    * Interpretation: Converting *BTC → ETH* costs 2 units
    * parent[*ETH*] = *BTC*
* *ETH → SOL*(-1): 
    * dist[*SOL*] = min(∞, 2 + (-1)) = 1
        * Interpretation: Path *BTC → ETH →SOL* has total cost 1
        * Key insight: Despite *ETH → SOL* being profitable (-1), total path still costs 1
    * parent[*SOL*] = *ETH*
* *ETH → ADA*(3): 
    * dist[*ADA*] = min(∞, 2 + 3) = 5
        * Interpretation: Path *BTC → ETH → ADA* costs 5
    * parent[*ADA*] = *ETH*
* *ETH → DOT*(4): 
    * dist[*DOT*] = min(∞, 2 + 4) = 6
        * Interpretation: Path *BTC → ETH → DOT* costs 6
    * parent[*DOT*] = *ETH*
* *SOL → DOT*(2): 
    * dist[*DOT*] = min(6, 1 + 2) = 3
        * **Improvement found** Path *BTC → ETH → SOL → DOT* costs only 3 vs previous 6
    * parent[*DOT*] = SOL (parent pointer updated)
* *DOT → ADA*(1): 
    * dist[*ADA*] = min(5, 3 + 1) = 4
        * **Improvement** Path *BTC → ETH → SOL → DOT → ADA* costs 4 vs previous 5
    * parent[*ADA*] = *DOT*

##### State after Iteration 1: 
**dist = {BTC:0, ETH:2, SOL:1, ADA:4, DOT:3}**<br>
Paths found: <br>
All paths of length ≤ 2 from *BTC*

##### Iteration 2 - Second Round: 
*Goal: Find all paths of length ≤ 2 from source*
Process all edges again (some may not improve distances):
* *BTC → ETH*(2): dist[*ETH*] = min(2, 0+2) = 2 (no change)
* *ETH → SOL*(-1): dist[*SOL*] = min(1, 2+(-1)) = 1 (no change)
* *ETH → ADA*(3): dist[*ADA*] = min(4, 2+3) = 4 (no change - current path is better)
* *ETH → DOT*(4): dist[*DOT*] = min(3, 2+4) = 3 (no change - current path via SOL is better)
* *SOL → DOT*(2): dist[*DOT*] = min(3, 1+2) = 3 (no change)
* *DOT → ADA*(1): dist[*ADA*] = min(4, 3+1) = 4 (no change)
* *ADA → DOT*(1): dist[*DOT*] = min(3, 4+1) = 3 (no change)
*Note: Assume we also have ADA → DOT edge with weight 1 for completeness*
##### State after Iteration 2: 
No changes - distances stabilized early
##### Iteration 3 & 4: 
Continue processing all edges, but no improvements found.<br>
Important step: <br>
**Negative Cycle Detection (Iteration 5):** 
Run one more iteration after |*V*| - 1 iterations. If ANY distance improves, a negative cycle exists.<br>
Suppose we discover a hidden edge *SOL → ETH*(-3) representing a special bonus trade.

##### Cycle Detection Iteration:
New edge *SOL* → *ETH*(-3): <br>
dist[*ETH*] = min(2, 1 + (-3)) = min(2, -2) = -2 <br>
An improvement is found and this indicates a negative cycle.<br>
The cycle is: <br>
***ETH → SOL → ETH*** 
with total weight (-1) + (-3) = -4 < 0
##### Negative Cycle Extraction Algorithm:

````python
def extract_negative_cycle(parent, improved_vertex):
    # Start from the improved vertex
    current = improved_vertex
    visited = set()
    
    # Walk back n steps to ensure we're in the cycle
    for _ in range(len(vertices)):
        current = parent[current]
    
    # Now extract the cycle
    cycle = []
    start = current
    while True:
        cycle.append(current)
        current = parent[current]
        if current == start:
            cycle.append(current)  	# Close the cycle
            break
    
    return cycle
````

#### Economic Interpretation of Negative Cycle 
If we find cycle *ETH* → *SOL* → *ETH* with total weight -4
* Convert *ETH* to *SOL*: gain factor e<sup>1</sup> ≈ 2.718
* Convert *SOL* back to *ETH*: gain factor e<sup>3</sup> ≈ 20.09
Total multiplicative gain: 2.718 × 20.09 ≈ 54.6
* Arbitrage opportunity: Start with 1 *ETH*, end with ~54.6 *ETH*

### Time Complexity: 
#### O(*V × E*)
* *V* - 1 iterations of relaxing E edges: O((*V* - 1) × *E*) = O(*V × E*)
* One additional iteration for cycle detection: O(*E*)
* Cycle extraction: O(*V*)
### Space Complexity: 
#### O(V) for distance and parent arrays
### Correctness Proof:
1. No negative cycles: After *V* - 1 iterations, all shortest paths are found (longest simple path has *V* - 1 edges)
2. Negative cycle exists: If we can still improve distances after *V* - 1 iterations, there must be a path of length ≥ *V*, which implies a cycle (by pigeonhole principle)
3. The cycle must be negative. Otherwise, we wouldn't improve distances by using it
##### Practical Considerations:
* Floating point precision: In real implementations, use tolerance for comparisons
* Multiple negative cycles: The algorithm found one but others may exist
* Inaccessible negative cycles: Only cycles reachable from the source are detected
##### Why Dijkstra's Fails Here: 
With the edge *ETH → SOL*(-1), Dijkstra's might process *ETH* (cost 2) and immediately finalize its distance. But later, if we find a path to *SOL* with cost < 1, we could improve *ETH*'s distance to 1 + (-1) = 0, violating Dijkstra's assumption that processed vertices have final distances.
