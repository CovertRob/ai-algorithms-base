## Coin Change Challenge
### Find the best combination of coins for a target amount

This section defines the helper functions for the Coin Change Challenge:

- **`format_coins(coins)`**: Formats coin denominations as e.g. 1¢ (penny), 5¢ (nickel), 10¢ (dime), 25¢ (quarter).
- **`format_combination(combination)`**: Shows counts with correct pluralization (e.g. `2 dimes, 3 pennies`).
- **`coin_change_dp(coins, amount)`**: Returns the minimum number of coins needed to make `amount` (in cents), or `-1` if impossible.
- **`get_coin_combination(coins, amount)`**: Returns a list of coins that form the optimal solution (or `[]` if impossible).

Notes:
- All amounts in the code are in cents. For example, 41 means $0.41.
- To use these functions, run the code cell in this section to load them. After loading, you can run the test examples in the test section to see results.

```python
from typing import List

# Names of common US coins
COIN_NAMES = {1: "penny", 5: "nickel", 10: "dime", 25: "quarter"}


def format_coins(coins: List[int]) -> str:
    """Format coin values with their names."""
    coin_items = []
    for coin in coins:
        name = COIN_NAMES.get(coin, f"${coin}")
        coin_items.append(f"{coin}¢ ({name})")
    return "[" + ", ".join(coin_items) + "]"


def format_combination(combination: List[int]) -> str:
    """Format coin combination with names."""
    if not combination:
        return "[]"
    
    coin_counts = {}
    for coin in combination:
        coin_counts[coin] = coin_counts.get(coin, 0) + 1
    
    formatted = []
    for coin in sorted(coin_counts.keys(), reverse=True):
        count = coin_counts[coin]
        name = COIN_NAMES.get(coin, f"${coin}")
        if count == 1:
            formatted.append(f"1 {name}")
        else:
            # Handle pluralization cases
            if name == "penny":
                plural = "pennies"
            else:
                plural = name if name.endswith("s") else name + "s"
            formatted.append(f"{count} {plural}")
    
    return ", ".join(formatted)


def coin_change_dp(coins: List[int], amount: int) -> int:
    """
    Return the minimum number of coins (denominations in `coins`, in cents)
    needed to make `amount` (in cents). Returns -1 if the amount cannot be formed.
    """
    if amount < 0:
        return -1
    # Use amount + 1 as "infinity" (greater than any possible coin count)
    dp = [amount + 1] * (amount + 1)
    dp[0] = 0

    for a in range(1, amount + 1):
        for coin in coins:
            if coin <= a and dp[a - coin] + 1 < dp[a]:
                dp[a] = dp[a - coin] + 1

    return dp[amount] if dp[amount] <= amount else -1


def get_coin_combination(coins: List[int], amount: int) -> List[int]:
    """Get the actual coins used to make the target amount (or [] if impossible)."""
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    parent = [-1] * (amount + 1)

    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i and dp[i - coin] + 1 < dp[i]:
                dp[i] = dp[i - coin] + 1
                parent[i] = coin

    if dp[amount] == float('inf'):
        return []

    result = []
    current = amount
    while current > 0:
        coin_used = parent[current]
        result.append(coin_used)
        current -= coin_used

    return sorted(result, reverse=True)
```

> coin denomination that is less than the current remaining amount.

> 20 + 1 + 1 + 1, while the optimal solution is 15+4+4.



## Conceptual Connection: Bellman-Ford and Dynamic Programming

Bellman-Ford computes the shortest cost to reach each node using the following dynamic programming recurrence:


$V(v) = \min_{(u, v) \in \text{Edges}} \left\{ V(u) + w(u, v) \right\}$

Where:
- $V(v)$ is the cost to reach node $v$
- $w(u, v)$ is the weight (time, cost, etc.) of the edge
- The algorithm performs $|V| - 1$ full passes to propagate costs optimally

This example models a delivery routing system with potential toll roads or time incentives.

```python
nodes = ['A (Warehouse)', 'B', 'C', 'D', 'E (Final drop-off)']

# Map node names to integer IDs for internal indexing
node_to_id = {name: i for i, name in enumerate(nodes)}
id_to_node = {i: name for name, i in node_to_id.items()}

# Edge list: (from_node, to_node, cost)
edges = [
    ('A (Warehouse)', 'B', 5),     # Warehouse to B: 5 min
    ('A (Warehouse)', 'C', 2),     # Warehouse to C: 2 min
    ('B', 'D', 3),                 # B to D: 3 min
    ('C', 'B', 1),                 # Shortcut from C to B: 1 min
    ('C', 'D', 10),                # Long route from C to D: 10 min
    ('D', 'E (Final drop-off)', -2),  # Toll road to E gives time benefit
]

def bellman_ford(edges, nodes, start_node, trace=False):
    """
    Bellman-Ford algorithm for shortest path from start_node.
    edges: list of (u, v, cost)
    nodes: list of node names
    start_node: string
    trace: whether to print edge relaxations
    """
    num_nodes = len(nodes)
    dist = {node: float('inf') for node in nodes}
    dist[start_node] = 0

    # Relax all edges |V| - 1 times
    for _ in range(num_nodes - 1):
        for u, v, cost in edges:
            if dist[u] + cost < dist[v]:
                dist[v] = dist[u] + cost
                if trace:
                    print(f"Relaxing edge {u} → {v}: new dist[{v}] = {dist[v]}")

    # Optional: detect negative-weight cycles
    for u, v, cost in edges:
        if dist[u] + cost < dist[v]:
            raise Exception("Negative-weight cycle detected!")

    return dist

# Run the delivery routing example
distances = bellman_ford(edges, nodes, start_node='A (Warehouse)', trace=True)

# Display final results
print("\nFinal Delivery Times from Warehouse:")
for node in nodes:
    print(f"{node}: {distances[node]} minutes")
```
