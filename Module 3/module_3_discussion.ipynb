## Is this Dynamic Programming?
1. Merge Sort
Merge splits an array into two halves and sorts each half recursively and then merges them back together as it traces back up the callstack. This is not a dynamic programming program, despite the fact that it uses recursion and breaks the problem into sub-problems, because none of the sub-problems have overlaps. There is not a means to further optimize merge sort by using techniques such as memoization because each sub-problem is entirely unique.
2. In Towers of Hanoi, you move *n* disks from peg A to peg C using peg B, obeying the size rule which is each disk has a distinct size and smaller disks must always be on top of larger disks. In Hanoi, the sub-problems are essentially standard recursion in that each subproblem is generated in a very structured way and there isn't meaningful recomputation of identical subproblems that would create a big savings if you memoized. The optimization here is to find the minimum number of moves, which is this problem is a closed form.
3. In Minimum coin change, you are given a set of coin denominations and a target amount and need to find the minimum number of coins needed to meeet that target amount. This is a good canidate for dynamic programming solutions because the subproblems overlap. The minimum amo unt of coins for amount x reuses the minimum amount of coins for x - 1. So the global optimal state is made up of optimal sub-problems.

## Climbing Stairs

You are climbing a staircase with n steps. You can climb either 1 step or 2 steps at a time.
How many distinct wars are there to reach the top

### Code examples:

```python
 
  # Top-Down (Memoized)
def climb_td(n, memo):
    if n in memo:
      return memo[n]
    memo[n] = climb_td(n-1, memo) + climb_td(n-2, memo)
    return memo[n]
```

```python
  # Bottom-Up (Tabulated)
def climb_bu(n):
    dp = [0] * (n+1)
    dp[0], dp[1] = 1, 1
    for i in range(2, n+1):
      dp[i] = dp[i-1] + dp[i-2]
    return dp[n]
```

1. Define the DP state.
    A general state definition can be `dp[n] = number of distinct ways to reach step n`
    However, there are two possible state transitions in this problem, with either taking 1 step or 2 steps for the next state. So we can define it more precisely as
    `dp[n] = dp[n-1] + dp[n-2]`, which is the recurrence definition.
2. Base cases.
    The base cases are when you just stand at the bottom and if there is only one way up from either of the states.
    `dp[0] = 1`
    `dp[1] = 1`
    `dp[2]` is not a true base case because it falls out of the recurrence
3. Explain why this problem has:
    - Optimal substructure
    It has optimal substructure because each next state is entirely determined by the previous one. Any valid way to reach step n must end with either a 1-step move from step n-1 or a 2-step move from step n-2. So the number of ways to reach step n is the sum of the number of ways to reach steps n-1 and n-2. So optimal solutions to those sub-problems must provide an optimal solution to our recurrently defined global problem.
    - Overlapping subproblems.
    The naive recursion of the staircase problem is extremely similar to the fibonacci naive recursion in that there are many repeated calculations across the recursive stack. Looking at the recurrence relation its pretty easy to see for stair cases greater than 3 steps there will be repeated calculations with optimization methods.
4. Comparing the implementations
    - What is shared between the two approaches?
    Both share the same state and recurrence by computing `dp[n] = dp[n-1] + dp[n-2]` and conceptually have the same bases cases.
    - What is the key difference in execution order?
    One is top down (memoized) and the other is bottom-up (tabulated). They just are opposite in the order by which they hit the base cases.
    - Which approach uses more memory, and why?
    Top-down (recursive) uses more memory because in addition to computing the values it must store the call stack each time. The tabulation just has to allocate an array of length n+1 and doesn't need the overhead of the call stack.
    - Which would you choose for large n, and why?
    I would definitely pick the tabulation method because there are recursion limits in Python in particular. It also has less overhead and can be implemented with constant extra memory.

### Bellman Update in Code

Given a weighted directed graph as a list of edges:

```python
edges = [
    ('A', 'B', 4),
    ('A', 'C', 2),
    ('C', 'B', -1),
    ('B', 'D', 2),
    ('C', 'D', 5)
  ] 
```

- Bellman equation for the value (distance) of node B
`d(v)=(u→v)min​{d(u)+w(u,v)}`
so for node b the equation is: `d(B) = min(d(A) + 4, d(C) - 1)`
- What the equation is computing
The equation says that the shortest distance to node B is the minimum of the following:
    - shortest distance to A plus the cost to go from A to B
    - shortest distance to C plus the cust to go from C to B
So Bellman-Ford checks every possible predecessor of B and asks if I come through this path, is that cheaper than what I currently believe is the most optimal.
- Completed function with one Bellman pass over all edges:

```python
def bellman_update(dist, edges):
    for u, v, w in edges:
        print(f"Edge {u} -> {v} (w = {w})")

        if dist[u] == float('inf'):
            print(f"  dist[{u}] = inf, no update")
            continue

        candidate = dist[u] + w
        print(f"  current dist[{v}] = {dist[v]}")
        print(f"  candidate = dist[{u}] + {w} = {candidate}")

        if candidate < dist[v]:
            dist[v] = candidate
            print(f"  updated dist[{v}] = {dist[v]}")
        else:
            print(f"  no update")

        print(f"  dist = {dist}\n")

    return dist

edges = [
    ('A', 'B', 4),
    ('A', 'C', 2),
    ('C', 'B', -1),
    ('B', 'D', 2),
    ('C', 'D', 5)
]

dist = {
    'A': 0,
    'B': float('inf'),
    'C': float('inf'),
    'D': float('inf')
}

bellman_update(dist, edges)
```
