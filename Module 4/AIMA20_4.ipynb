## Local Search and Optimization Problems
- Local search algorithms operate by searching from a start state to neighboring states, without keeping track of the paths, nor the set of states that have been reached
- Can solve **optimization problems** in which aim is to find the best state according to the **objective function**

### Hill-Climbing Search
- if elevation corresponds to an objective function -- aim is to find highest peak -- global maximum
- if elevation corresponds to cost -- find lowest valley -- global minimum -- **gradient descent**
- hill-climbing keeps track of one current state and on each iteration moves to neighboring state w/ highest value
    - Aka *greedy local search*
- local search algorithms can get lost in **local maxima** , **ridges**, and **plateaus**
    - some optimization methods include **stochastic hill climbing** and **first choice hill-climbing**
- **Success of hill-climbing depends very much on the state of the shape of the state-space landscape**
### Simulated Annealing
- instead of picking best move, it picks a random move
- if moves improves situation -> accepted
- otherwise acceps the move with some probability less than 1
- The probability decreases exponentially with the "badness" of the move - the amount $\delta E$ by which the evaluation worsened
- The probability also decreases as the temperate $T$ goes down: "bad" moves are more likely to be allowed at the start when $T$ is high, and they become more unlikely as $T$ decreases.
- If the schedule lowers $T$ to 0 slowly enough, then a property of the Boltzmann distributiohn $e^\deltaE/T$ is that all the probability is concentrated on the global maxima, which the algorithm will find with probability approaching 1
- Schedule input determines the value of T (temperature) as a function of time
### Local Beam Search
- Use variant stochastic beam search to spread out searches among diversified k-states
- In a local beam search, useful information is passed among the parallel search threads
