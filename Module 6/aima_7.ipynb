# Notes from AIMA Chapter 7 Logical Agents

## 7.1 Knowledge-Based Agents
- **Knowledge-based agents**: Agents that use a process of reasoning over an internal **representation** of knowledge to decide which actions to take.
- **Reasoning**: The internal process of manipulating representations to derive new knowledge or make decisions.
- Knowledge-based agents are more flexible than problem-solving agents because they can combine information for myriad purposes and adapt to environment changes by updating knowledge.
- **Knowledge base** (**KB**): The central component of the agent; a set of **sentences**.
- **Sentence**: A technical term for an individual assertion about the world expressed in a **knowledge representation language**.
- **Axiom**: A sentence taken as given without being derived from other sentences.
- **TELL**: Operation to add new sentences to the knowledge base.
- **ASK**: Operation to query what is known; the answer must "follow" from what has been TELLed.
- **Inference**: The process of deriving new sentences from old ones; it ensures the answer to an ASK query is grounded in previous information.
- **Background knowledge**: The initial knowledge base containing general facts about the world or the specific task environment.
- **Algorithm: KB-AGENT**
```text
function KB-AGENT(percept) returns an action
  persistent: KB, a knowledge base
              t, a counter, initially 0, indicating time
  TELL(KB, MAKE-PERCEPT-SENTENCE(percept, t))
  action <- ASK(KB, MAKE-ACTION-QUERY(t))
  TELL(KB, MAKE-ACTION-SENTENCE(action, t))
  t <- t + 1
  return action
```
- Each call to the agent program triggers three steps: TELLing the KB what is perceived, ASKing for an action, and TELLing the KB which action was taken.
- Interface functions hide representation details: `MAKE-PERCEPT-SENTENCE` (constructs assertion of perception at time $t$), `MAKE-ACTION-QUERY` (asks for the best action at time $t$), and `MAKE-ACTION-SENTENCE` (asserts that a specific action was executed).
- **Knowledge level**: The highest level of agent description, specifying only what the agent knows and its goals to determine behavior.
- **Implementation level**: The level of description concerning how knowledge is physically stored (e.g., linked lists, strings, or neuronal signals).
- **Declarative**: Approach to system building where an agent is constructed by TELLing it sentences one by one until it knows how to operate.
- **Procedural**: Approach that encodes desired behaviors directly into program code.

## Task Environment Specification (PEAS)
- **PEAS**: Acronym for **Performance, Environment, Actuators, Sensors**.
- The first step in designing any agent, including knowledge-based ones, must be to specify the task environment as fully as possible using PEAS.
- This step is analogous to the first step of **Knowledge Engineering**: identifying the range of questions the KB will support and the facts available.
- Example: **Wumpus World PEAS**
- **Performance Measure**: $+1000$ for climbing out with gold, $-1000$ for death, $-1$ per action, $-10$ for using the arrow.
- **Environment**: $4 \times 4$ grid of rooms; contains a Wumpus, pits, and gold.
- **Actuators**: $Forward$, $TurnLeft$, $TurnRight$, $Grab$, $Shoot$, $Climb$.
- **Sensors**: $Stench$ (near Wumpus), $Breeze$ (near pit), $Glitter$ (at gold), $Bump$ (wall), $Scream$ (Wumpus death).

## 7.3 Logic
- **Logic**: Fundamental concepts of representation and reasoning that are independent of specific forms.
- **Syntax**: The rules specifying all sentences that are well formed in the representation language (e.g., $x+y=4$ is well-formed).
- **Semantics**: Defines the "meaning" or truth of sentences with respect to each possible world.
- **Possible world**: A potential real environment an agent might inhabit; sentences must be true or false in these worlds.
- **Model**: A mathematical abstraction of a possible world that assigns a fixed truth value (true or false) to every relevant sentence.
- **Satisfaction**: If a sentence $\alpha$ is true in a specific model $m$, we say that $m$ satisfies $\alpha$ (or $m$ is a **model of** $\alpha$).
- $M(\alpha)$ is the notation for the set of all models of the sentence $\alpha$.
- **Entailment**: The logical relationship where one sentence follows from another; written as $\alpha \models \beta$.
- Formal definition of entailment: $\alpha \models \beta$ if and only if in every model in which $\alpha$ is true, $\beta$ is also true.
- Set-theoretic definition: $\alpha \models \beta$ if and only if the set of models of $\alpha$ is a subset of the models of $\beta$:
$$ \alpha \models \beta \iff M(\alpha) \subseteq M(\beta) $$
- Directionality: If $\alpha \models \beta$, then $\alpha$ is a stronger assertion than $\beta$ because it rules out more possible worlds.
- **Logical inference**: The practical process of deriving conclusions from a knowledge base ($KB$).
- **Model checking**: An inference algorithm that performs an enumeration of all possible models to verify if $M(KB) \subseteq M(\alpha)$.
- Inference notation: $KB \vdash_i \alpha$ is read as "$\alpha$ is derived from $KB$ by algorithm $i$."
- **Sound** (**truth-preserving**): An inference algorithm is sound if it derives only sentences that are actually entailed by the $KB$.
- **Completeness**: An inference algorithm is complete if it can derive any sentence that is entailed by the $KB$.
- Correspondence: A sound inference procedure ensures that if the $KB$ is true in the real world, any sentence $\alpha$ derived from it is also true.
- **Grounding**: The connection between logical reasoning processes (internal syntax) and the real environment, established via sensors or learning.

## 7.4 Propositional Logic: A Very Simple Logic
- **Propositional logic**: A simple logic based on propositions that can be true or false, providing a foundation for understanding representation and inference.
## 7.4.1 Syntax
- The syntax defines allowable sentences. **Atomic sentences** consist of a single **proposition symbol** (e.g., $P, Q, W_{1,3}$).
- Symbols $True$ and $False$ have fixed meanings (always true and always false, respectively).
- **Complex sentences** are built from simpler sentences using parentheses and **logical connectives**.
- **Negation** ($\neg$): Operates on a sentence to produce its negation (e.g., $\neg W_{1,3}$ means "not $W_{1,3}$").
- **Literal**: An atomic sentence (**positive literal**) or a negated atomic sentence (**negative literal**).
- **Conjunction** ($\wedge$): The "and" connective; parts are called **conjuncts** (e.g., $W_{1,3} \wedge P_{3,1}$).
- **Disjunction** ($\vee$): The "or" connective; parts are called **disjuncts** (e.g., $W_{1,3} \vee P_{3,1}$).
- **Implication** ($\Rightarrow$): The "implies" or conditional connective; relates a **premise** (or **antecedent**) to a **conclusion** (or **consequent**).
- Implications are also known as **rules** or **if-then statements**.
- **Biconditional** ($\Leftrightarrow$): The "if and only if" connective (e.g., $W_{1,3} \Leftrightarrow \neg W_{2,2}$).
- Operator Precedence (highest to lowest): $\neg, \wedge, \vee, \Rightarrow, \Leftrightarrow$.
## 7.4.2 Semantics
- Semantics define rules for determining the truth of a sentence with respect to a particular **model**.
- **Truth value**: In propositional logic, a **model** sets the truth value (true or false) for every proposition symbol.
- Recursive rules for truth of complex sentences in model $m$:
- $\neg P$ is true iff $P$ is false in $m$.
- $P \wedge Q$ is true iff $P$ and $Q$ are both true in $m$.
- $P \vee Q$ is true iff $P$ or $Q$ is true in $m$.
- $P \Rightarrow Q$ is true unless $P$ is true and $Q$ is false in $m$.
- $P \Leftrightarrow Q$ is true iff $P$ and $Q$ are both true or both false in $m$.
- **Truth table**: A matrix specifying the truth value of a complex sentence for all possible assignments to its components.
## 7.4.3 A Simple Knowledge Base
- A KB for the wumpus world uses symbols like $P_{x,y}$ (pit at $[x,y]$) and $B_{x,y}$ (breeze at $[x,y]$).
- Rules are expressed using connectives, such as $R_2: B_{1,1} \Leftrightarrow (P_{1,2} \vee P_{2,1})$ (a square is breezy iff a neighbor has a pit).
## 7.4.4 A Simple Inference Procedure
- Goal: Determine if $KB \models \alpha$ (does the KB entail sentence $\alpha$?).
- **Model checking**: An inference approach that enumerates all possible models to verify $\alpha$ is true in all models where $KB$ is true.
- For $n$ proposition symbols, there are $2^n$ models; time complexity is $O(2^n)$.
- Propositional entailment is **co-NP-complete**.
- **Algorithm: TT-ENTAILS?**
```text
function TT-ENTAILS?(KB, α) returns true or false
  inputs: KB, the knowledge base, a sentence in propositional logic
          α, the query, a sentence in propositional logic
  symbols <- a list of the proposition symbols in KB and α
  return TT-CHECK-ALL(KB, α, symbols, {})

function TT-CHECK-ALL(KB, α, symbols, model) returns true or false
  if EMPTY?(symbols) then
    if PL-TRUE?(KB, model) then return PL-TRUE?(α, model)
    else return true // when KB is false, always return true
  else
    P <- FIRST(symbols); rest <- REST(symbols)
    return (TT-CHECK-ALL(KB, α, rest, model ∪ {P = true})
            and
            TT-CHECK-ALL(KB, α, rest, model ∪ {P = false}))
```
- The algorithm is sound and complete because it directly implements the definition of entailment over a finite model space.

## 7.5 Propositional Theorem Proving
- **Theorem proving**: The process of determining entailment by applying inference rules directly to sentences in the knowledge base to construct a proof without consulting models.
- **Logical equivalence**: Two sentences $\alpha$ and $\beta$ are equivalent if they are true in the same set of models, written $\alpha \equiv \beta$.
- Alternative definition: $\alpha \equiv \beta$ if and only if $\alpha \models \beta$ and $\beta \models \alpha$.
- **Validity**: A sentence is valid if it is true in all possible models (also called **tautologies**).
- **Deduction theorem**: For any sentences $\alpha$ and $\beta$, $\alpha \models \beta$ if and only if the sentence $(\alpha \Rightarrow \beta)$ is valid.
- **Satisfiability**: A sentence is satisfiable if it is true in, or satisfied by, at least one model.
- **SAT**: The problem of determining the satisfiability of sentences in propositional logic; it was the first problem proved to be NP-complete.
- Relationship between validity and satisfiability: $\alpha$ is valid iff $\neg \alpha$ is unsatisfiable.
- Relationship to entailment: $KB \models \alpha$ iff $(KB \wedge \neg \alpha)$ is unsatisfiable.
- **Reductio ad absurdum**: Proof technique (also **proof by refutation** or **proof by contradiction**) where one assumes a sentence to be false and shows it leads to a contradiction.

## 7.5.1 Inference and Proofs
- **Inference rules**: Standard patterns of sound inference applied to derive a **proof** (a chain of conclusions leading to a goal).
- **Modus Ponens**: Written as $\frac{\alpha \Rightarrow \beta, \alpha}{\beta}$; if an implication and its premise are given, the conclusion is inferred.
- **And-Elimination**: Written as $\frac{\alpha \wedge \beta}{\alpha}$; from a conjunction, any of the conjuncts can be inferred.
- All logical equivalences (e.g., De Morgan’s laws) can be used as sound inference rules.
- **Monotonicity**: A property of logical systems stating that the set of entailed sentences can only increase as information is added: if $KB \models \alpha$, then $(KB \wedge \beta) \models \alpha$.

## 7.5.2 Proof by Resolution
- **Resolution**: A single, complete inference rule that yields a complete inference algorithm when coupled with a complete search algorithm.
- **Resolvent**: The new clause produced by resolving two clauses containing **complementary literals**.
- **Unit resolution**: $\frac{\ell_1 \vee \dots \vee \ell_k, m}{\ell_1 \vee \dots \vee \ell_{i-1} \vee \ell_{i+1} \vee \dots \vee \ell_k}$ where $m$ and $\ell_i$ are complementary.
- **Clause**: A disjunction of literals; a **unit clause** consists of just one literal.
- **Resolution rule**: $\frac{\ell_1 \vee \dots \vee \ell_k, m_1 \vee \dots \vee m_n}{\ell_1 \vee \dots \vee \ell_{i-1} \vee \ell_{i+1} \vee \dots \vee \ell_k \vee m_1 \vee \dots \vee m_{j-1} \vee m_{j+1} \vee \dots \vee m_n}$ where $\ell_i$ and $m_j$ are complementary.
- **Factoring**: The removal of multiple copies of the same literal in a resulting clause.
- **Conjunctive normal form (CNF)**: A representation where a sentence is a conjunction of clauses.
- Conversion to CNF: 1. Eliminate $\Leftrightarrow$; 2. Eliminate $\Rightarrow$; 3. Move $\neg$ inwards (De Morgan/Double negation); 4. Distribute $\vee$ over $\wedge$.
- **Algorithm: PL-RESOLUTION**
```text
function PL-RESOLUTION(KB, α) returns true or false
  inputs: KB, the knowledge base, a sentence in propositional logic
          α, the query, a sentence in propositional logic
  clauses <- the set of clauses in the CNF representation of KB ∧ ¬α
  new <- {}
  while true do
    for each pair of clauses Ci, Cj in clauses do
      resolvents <- PL-RESOLVE(Ci, Cj)
      if resolvents contains the empty clause then return true
      new <- new ∪ resolvents
    if new ⊆ clauses then return false
    clauses <- clauses ∪ new
```
- **Resolution closure**: The set $RC(S)$ of all clauses derivable by repeated application of the resolution rule to set $S$.
- **Ground resolution theorem**: If a set of clauses is unsatisfiable, then its resolution closure contains the empty clause.

## 7.5.3 Horn Clauses and Definite Clauses
- **Definite clause**: A disjunction of literals of which exactly one is positive.
- **Horn clause**: A disjunction of literals of which at most one is positive (includes definite clauses and **goal clauses**).
- **Goal clauses**: Horn clauses with no positive literals.
- Implication form: A definite clause $( \neg L_1 \vee \dots \vee \neg L_n \vee Head )$ is written $( L_1 \wedge \dots \wedge L_n \Rightarrow Head )$.
- **Body**: The conjunction of positive literals in the premise of the implication.
- **Head**: The single positive literal forming the conclusion.
- **Fact**: A sentence consisting of a single positive literal.
- Deciding entailment with Horn clauses is efficient and takes time linear in the size of the KB.

## 7.5.4 Forward and Backward Chaining
- **Forward-chaining**: A **data-driven** reasoning process that starts from known facts and triggers rules whose premises are satisfied, adding conclusions to the KB.
- **Algorithm: PL-FC-ENTAILS?**
```text
function PL-FC-ENTAILS?(KB, q) returns true or false
  inputs: KB, the knowledge base, a set of propositional definite clauses
          q, the query, a proposition symbol
  count <- a table, where count[c] is initially the number of symbols in clause c's premise
  inferred <- a table, where inferred[s] is initially false for all symbols
  queue <- a queue of symbols, initially symbols known to be true in KB
  while queue is not empty do
    p <- POP(queue)
    if p = q then return true
    if inferred[p] = false then
      inferred[p] <- true
      for each clause c in KB where p is in c.PREMISE do
        decrement count[c]
        if count[c] = 0 then add c.CONCLUSION to queue
  return false
```
- **Backward-chaining**: A **goal-directed** reasoning process that works backward from the query, finding implications that conclude the query and attempting to prove their premises.

## 7.6 Effective Propositional Model Checking
- This section describes two families of efficient algorithms for general propositional inference based on model checking: one based on backtracking search and one on local hill-climbing search.
- These algorithms check for satisfiability—the **SAT** problem.
- Testing entailment $KB \models \alpha$ is equivalent to checking the unsatisfiability of $(KB \wedge \neg \alpha)$.
- Many combinatorial problems can be reduced to checking the satisfiability of a propositional sentence, making SAT algorithms highly important for handling complexity.
## 7.6.1 A complete backtracking algorithm
- **Davis–Putnam algorithm** (**DPLL**): A complete, recursive, depth-first enumeration of possible models for sentences in CNF.
- DPLL utilizes three key heuristics to improve upon basic truth-table enumeration:
- **Early termination**: The algorithm stops checking a branch if the sentence must already be true (e.g., one literal in every clause is true) or false (e.g., all literals in one clause are false).
- **Pure symbol heuristic**: A **pure symbol** is one that always appears with the same sign in all unsatisfied clauses. If a model exists, one exists with pure symbols assigned to make their literals true.
- **Unit clause heuristic**: Initially refers to clauses with one literal; in DPLL, it includes clauses where all literals but one have been assigned false.
- **Unit propagation**: The process where assigning a value to a unit clause creates new unit clauses in a cascade of forced assignments.
- **Algorithm: DPLL-SATISFIABLE?**
```text
function DPLL-SATISFIABLE?(s) returns true or false
  inputs: s, a sentence in propositional logic
  clauses <- the set of clauses in the CNF representation of s
  symbols <- a list of the proposition symbols in s
  return DPLL(clauses, symbols, {})
function DPLL(clauses, symbols, model) returns true or false
  if every clause in clauses is true in model then return true
  if some clause in clauses is false in model then return false
  P, value <- FIND-PURE-SYMBOL(symbols, clauses, model)
  if P is non-null then return DPLL(clauses, symbols - P, model ∪ {P=value})
  P, value <- FIND-UNIT-CLAUSE(clauses, model)
  if P is non-null then return DPLL(clauses, symbols - P, model ∪ {P=value})
  P <- FIRST(symbols); rest <- REST(symbols)
  return DPLL(clauses, rest, model ∪ {P=true}) or DPLL(clauses, rest, model ∪ {P=false})
```
- Advanced solvers include additional scaling enhancements:
- **Component analysis**: Partitioning clauses into disjoint subsets (**components**) that share no unassigned variables to solve them independently.
- **Variable and value ordering**: Using heuristics like the degree heuristic (choosing the variable that appears most frequently in unsatisfied clauses).
- **Intelligent backtracking**: Backing up past the most recent decision to the actual point of conflict; often involves conflict clause learning.
- **Random restarts**: Starting search over with different random choices to reduce time variance when progress appears to stall.
- **Clever indexing**: Implementing dynamic data structures to quickly identify clauses containing specific literals.
## 7.6.2 Local search algorithms
- Local search algorithms use a complete-state formulation, assigning a value to every variable and flipping them to resolve conflicts.
- Evaluation function: Typically counts the number of unsatisfied clauses, seeking a state with a score of zero.
- **WALKSAT**: A stochastic local search algorithm that balances greedy "min-conflicts" moves with "random walk" moves to avoid local minima.
- **Algorithm: WALKSAT**
```text
function WALKSAT(clauses, p, max_flips) returns a satisfying model or failure
  inputs: clauses, a set of clauses in propositional logic
          p, the probability of choosing a "random walk" move (typically around 0.5)
          max_flips, number of value flips allowed before giving up
  model <- a random assignment of true/false to the symbols in clauses
  for i = 1 to max_flips do
    if model satisfies clauses then return model
    clause <- a randomly selected clause from clauses that is false in model
    if RANDOM(0, 1) <= p then
      flip the value in model of a randomly selected symbol from clause
    else flip whichever symbol in clause maximizes the number of satisfied clauses
  return failure
```
## 7.6.3 The landscape of random SAT problems
- Problem difficulty is characterized by the ratio $m/n$, where $m$ is the number of clauses and $n$ is the number of symbols.
- Underconstrained problems (low $m/n$) are easy because models are abundant; overconstrained problems (high $m/n$) are easy because contradictions are found quickly.
- **Satisfiability threshold conjecture**: Theory stating that for every $k \ge 3$, there is a threshold ratio $r_k$ where the probability of satisfiability drops sharply from 1 to 0 as $n \to \infty$.
- For 3-SAT, this threshold is empirically found at $m/n \approx 4.3$.
- The most difficult problem instances are located precisely at this threshold value.

## 7.7 Agents Based on Propositional Logic
- This section combines propositional logic concepts to construct wumpus world agents that deduce the world state from percept history, perform state estimation, and construct plans.
## 7.7.1 The current state of the world
- The knowledge base (KB) consists of axioms (general physics of the world) and percept sentences from experience.
- Core axioms include the starting square being safe ($\neg P_{1,1}$, $\neg W_{1,1}$) and definitions for breezes/stenches: $B_{1,1} \Leftrightarrow (P_{1,2} \vee P_{2,1})$.
- Wumpus existence is defined by saying there is at least one ($W_{1,1} \vee W_{1,2} \vee \dots \vee W_{4,4}$) and at most one (for every pair of locations, at least one is wumpus-free).
- **Fluent**: An aspect of the world that changes over time, also called a state variable.
- **Atemporal variable**: A symbol representing a permanent aspect of the world that does not require a time superscript.
- Propositions are indexed by time $t$ to avoid contradictions (e.g., $L_{1,1}^0$ for location at time 0).
- **Effect axiom**: Specifies the outcome of an action at the next time step (e.g., $L_{1,1}^0 \wedge FacingEast^0 \wedge Forward^0 \Rightarrow L_{2,1}^1$).
- **Frame problem**: The challenge of representing all the things that do *not* change when an action is taken.
- **Frame axiom**: An axiom that explicitly asserts a proposition remains unchanged under a specific action.
- **Representational frame problem**: The inefficiency where a world with $m$ actions and $n$ fluents requires $O(mn)$ frame axioms.
- **Locality**: The property that each action typically changes only a small number $k$ of fluents.
- **Inferential frame problem**: The difficulty of projecting the results of a $t$-step plan in time proportional to $t$ rather than the number of fluents.
- **Successor-state axiom**: Solves the frame problem by focusing on fluents; for each fluent $F$, the axiom defines its truth at $t+1$ based on actions and state at $t$.
- Successor-state axiom schema: $$ F^{t+1} \Leftrightarrow \text{ActionCausesF}^t \vee (F^t \wedge \neg \text{ActionCausesNotF}^t) $$.
- **Qualification problem**: The impossibility of defining every unusual exception that might cause an action to fail.
## 7.7.2 A hybrid agent
- **Hybrid agent**: An agent that combines propositional logic (for state inference) with problem-solving search (A* for path planning).
- It updates the KB with temporal "physics" at each step and ASKs the KB which squares are safe or unvisited.
- Plan priority: 1. Grab gold and climb out; 2. Visit closest unvisited safe square; 3. Shoot wumpus if arrow exists; 4. Risk a non-unsafe square; 5. Retreat.
```text
function HYBRID-WUMPUS-AGENT(percept) returns an action
  persistent: KB, a knowledge base, initially the atemporal "wumpus physics"
              t, a counter, initially 0, indicating time
              plan, an action sequence, initially empty
  TELL(KB, MAKE-PERCEPT-SENTENCE(percept, t))
  TELL the KB the temporal "physics" sentences for time t
  safe <- {[x,y] : ASK(KB, OK^t_x,y) = true}
  if ASK(KB, Glitter^t) = true then
    plan <- [Grab] + PLAN-ROUTE(current, {}, safe) + [Climb]
  if plan is empty then
    unvisited <- {[x,y] : ASK(KB, L^t'_x,y) = false for all t' <= t}
    plan <- PLAN-ROUTE(current, unvisited ∩ safe, safe)
  if plan is empty and ASK(KB, HaveArrow^t) = true then
    possible_wumpus <- {[x,y] : ASK(KB, ¬W_x,y) = false}
    plan <- PLAN-SHOT(current, possible_wumpus, safe)
  if plan is empty then // no choice but to take a risk
    not_unsafe <- {[x,y] : ASK(KB, ¬OK^t_x,y) = false}
    plan <- PLAN-ROUTE(current, unvisited ∩ not_unsafe, safe)
  if plan is empty then
    plan <- PLAN-ROUTE(current, {}, safe) + [Climb]
  action <- POP(plan)
  TELL(KB, MAKE-ACTION-SENTENCE(action, t))
  t <- t + 1
  return action
```
## 7.7.3 Logical state estimation
- As history grows, the computational cost of `ASK` becomes proportional to the agent's life length.
- **Caching**: Saving inference results to build on for the next time step.
- **State estimation**: The process of updating the belief state as new percepts arrive.
- **Belief state**: A logical sentence representing the set of all possible current states of the world.
- Exact state estimation with logical formulas can require exponential size relative to the number of symbols.
- **Conservative approximation**: Representing the belief state as a 1-CNF formula (conjunction of literals) to maintain efficiency while losing some interaction information.
## 7.7.4 Making plans by propositional inference
- Planning can be performed by presenting a SAT solver with a sentence containing the initial state, successor-state axioms, and a goal assertion.
- **SATPLAN**: A procedure that tries each possible number of steps $t$ until it finds a satisfying model, ensuring the shortest plan is found.
```text
function SATPLAN(init, transition, goal, T_max) returns solution or failure
  inputs: init, transition, goal, constitute a description of the problem
          T_max, an upper limit for plan length
  for t = 0 to T_max do
    cnf <- TRANSLATE-TO-SAT(init, transition, goal, t)
    model <- SAT-SOLVER(cnf)
    if model is not null then
      return EXTRACT-SOLUTION(model)
  return failure
```
- **Precondition axioms**: Required for SATPLAN to prevent models with impossible actions (e.g., $Shoot^t \Rightarrow HaveArrow^t$).
- **Action exclusion axioms**: Required to prevent the solver from proposing multiple simultaneous actions (e.g., $\neg A_i^t \vee \neg A_j^t$).
- Propositional logic does not scale to unbounded environments because it cannot concisely represent universal patterns across all time and space.
