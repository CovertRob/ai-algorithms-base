## Question 1: Autonomous Vehicle as an AI System

### Introduction to Autonomous Vehicles as Complex AI Systems

Autonomous vehicles represent one of the most sophisticated applications of artificial intelligence in modern engineering, integrating multiple algorithmic systems to achieve safe, efficient navigation in dynamic environments. According to the lecture materials on Algorithms and AI, these systems exemplify the complete perception-reasoning-action-feedback loop that characterizes intelligent systems (Chen, 2026). The autonomous vehicle architecture demonstrates how different algorithmic components must work in concert to process sensory information, make intelligent decisions, execute actions, and continuously learn from environmental feedback. Understanding this system architecture provides insights into how complex AI systems orchestrate multiple specialized algorithms to solve real-world problems.

### Perception Stage: Sensing and Understanding the Environment

The perception stage serves as the foundation of autonomous vehicle intelligence, responsible for gathering and interpreting information from the physical world. At this stage, the vehicle deploys multiple sensor systems including LiDAR (Light Detection and Ranging), radar, cameras, ultrasonic sensors, and GPS receivers to create a comprehensive understanding of the environment. Computer vision algorithms process camera feeds to identify lane markings, traffic signs, traffic lights, pedestrians, and other vehicles through deep learning models such as convolutional neural networks. Simultaneously, LiDAR systems generate three-dimensional point clouds representing the vehicle's surroundings, which are processed through point cloud segmentation algorithms to distinguish between different objects and surfaces. Sensor fusion algorithms integrate data from multiple sources to create a unified environmental model, compensating for individual sensor limitations and increasing reliability. The perception system must also perform object detection and tracking, maintaining consistent identification of moving objects across frames to predict their trajectories. Additionally, localization algorithms combine GPS data with visual odometry and simultaneous localization and mapping techniques to determine the vehicle's precise position within a high-definition map with centimeter-level accuracy.

### Reasoning Stage: Decision Making and Planning

The reasoning stage transforms perceptual data into actionable decisions through sophisticated planning and decision-making algorithms. This stage encompasses several interconnected components that operate at different temporal scales. Path planning algorithms utilize graph search methods such as A-star or Rapidly-exploring Random Trees to determine optimal routes from the current location to the destination, considering factors like traffic conditions, road restrictions, and efficiency. Behavioral planning systems make high-level decisions about driving maneuvers, such as whether to change lanes, yield to other vehicles, or navigate intersections, using decision trees, finite state machines, or reinforcement learning policies. Motion planning algorithms compute detailed trajectories that the vehicle should follow, optimizing for safety, comfort, and traffic rule compliance while avoiding obstacles. These systems must evaluate multiple potential trajectories simultaneously, predicting how other road users might behave and selecting actions that minimize risk. The reasoning stage also incorporates traffic rule reasoning, ensuring that planned actions comply with traffic regulations, road signs, and cultural driving norms specific to the operating region.

### Action Stage: Executing Decisions Through Vehicle Control

The action stage translates high-level decisions into precise physical control of the vehicle's actuators. Control algorithms continuously adjust steering angle, throttle position, and brake pressure to follow the planned trajectory with high precision. Proportional-Integral-Derivative controllers or more advanced model predictive control algorithms ensure smooth execution while compensating for vehicle dynamics, road conditions, and environmental disturbances such as wind or road gradient. The control system must balance multiple objectives simultaneously, including following the desired path, maintaining passenger comfort, ensuring stability, and responding to unexpected situations. Low-level actuator control interfaces with the vehicle's mechanical systems, translating algorithmic commands into actual movements. The action stage also manages coordination between different control systems, ensuring that steering, acceleration, and braking work harmoniously. Safety monitoring systems at this stage verify that commanded actions fall within safe operating parameters and can trigger emergency responses if dangerous conditions are detected.

### Feedback Stage: Learning and Adaptation

The feedback stage completes the loop by monitoring the outcomes of actions and enabling continuous improvement of the system. Sensor feedback provides immediate information about the vehicle's actual state, including position, velocity, acceleration, and orientation, which is compared against intended states to compute control errors. This real-time feedback enables closed-loop control, where the system continuously corrects deviations from planned behavior. Performance monitoring algorithms track metrics such as passenger comfort, energy efficiency, and adherence to traffic rules, identifying areas for system improvement. Machine learning components use feedback to refine their models, improving perception accuracy, prediction quality, and decision-making effectiveness over time. Anomaly detection systems identify unusual situations or sensor failures, triggering appropriate safety responses. Data logging systems record sensor data, decisions, and outcomes for offline analysis, enabling engineers to identify edge cases and improve algorithms through simulation and retraining. The feedback stage also includes validation mechanisms that assess whether the vehicle's behavior aligns with safety requirements and expected performance standards.

### System Integration and Real-Time Orchestration

The complete autonomous vehicle system demonstrates how these four stages must operate in real-time coordination, with information flowing continuously through the perception-reasoning-action-feedback loop. The entire cycle typically operates at frequencies ranging from 10 to 100 Hz depending on the subsystem, requiring careful software architecture and computational resource management. Middleware systems facilitate communication between components, ensuring that data flows efficiently while meeting timing constraints. The integration of these stages reveals the complexity of AI systems, where individual algorithmic components must not only perform their designated tasks effectively but also work cohesively within a larger system architecture to achieve the overarching goal of safe autonomous navigation.

## Question 2: Algorithm Comparison - Fibonacci Computation

### Introduction to Fibonacci Algorithms and Complexity Analysis

The Fibonacci sequence represents a fundamental problem in computer science that reveals how algorithmic design choices dramatically impact computational efficiency. Named after the Italian mathematician Leonardo Fibonacci, this sequence begins with 0 and 1, with each subsequent number being the sum of the two preceding ones (0, 1, 1, 2, 3, 5, 8, 13, 21, ...). While multiple algorithms can compute the same result, their performance characteristics differ dramatically, making Fibonacci computation an ideal case study for understanding time and space complexity analysis. As discussed in the lecture materials on algorithm complexity analysis, evaluating algorithms requires systematic analysis of their resource requirements as input size grows (Chen, 2026). This question examines two contrasting approaches: naive recursion and dynamic programming, demonstrating how algorithmic sophistication can transform an exponential-time problem into a linear-time solution.

### Pseudocode A: Naive Recursion - Analysis

The naive recursive algorithm implements the mathematical definition of Fibonacci numbers directly, using the recurrence relation F(n) = F(n-1) + F(n-2) with base cases F(0) = 0 and F(1) = 1. This approach appears elegant and closely mirrors the mathematical definition, making it intuitively appealing. However, this simplicity conceals severe computational inefficiency. The time complexity of this algorithm is exponential, specifically O(2^n), because each function call generates two additional calls, creating a binary tree of recursive calls. For example, computing F(5) requires computing both F(4) and F(3), but F(4) itself requires computing F(3) and F(2), resulting in redundant computation of F(3). This redundant work grows explosively as n increases. More precisely, the number of function calls follows the recurrence T(n) = T(n-1) + T(n-2) + 1, which grows asymptotically as the golden ratio raised to the power n, approximately 1.618^n. The space complexity is O(n) due to the maximum depth of the recursion stack, which equals n in the worst case when the algorithm follows the leftmost branch of the recursion tree down to the base case. While the space requirement is linear, the exponential time complexity makes this algorithm impractical for even moderately large values of n.

### Pseudocode B: Dynamic Programming - Analysis  

The dynamic programming algorithm transforms the problem by recognizing and eliminating the redundant computations that plague the naive approach. Instead of recursively recomputing the same Fibonacci numbers multiple times, this algorithm builds the solution bottom-up, computing each Fibonacci number exactly once and storing it in an array F for future reference. The algorithm initializes F[0] = 0 and F[1] = 1, then iteratively computes F[i] = F[i-1] + F[i-2] for i from 2 to n. This approach exhibits linear time complexity O(n) because it performs exactly one addition operation for each value from 2 to n, resulting in precisely n-1 additions total. The space complexity is also O(n) because the algorithm maintains an array of n+1 elements to store all computed Fibonacci numbers from F[0] to F[n]. Importantly, the space complexity could be further optimized to O(1) by recognizing that only the two most recent Fibonacci numbers need to be stored at any time, though the pseudocode presented uses O(n) space. The dynamic programming approach exemplifies a fundamental principle in algorithm design: trading space for time by storing intermediate results to avoid redundant computation, a technique called memoization when applied to recursion or tabulation when implemented iteratively as shown here.

### Comparison for n = 40: Practical Differences

When computing F(40), the differences between these algorithms become starkly apparent in practice. The naive recursive algorithm would require approximately 2^40 ≈ 1.1 trillion function calls, specifically 331,160,281 calls based on the Fibonacci sequence's growth rate. On a modern computer performing one billion operations per second, this would require several minutes to complete. In contrast, the dynamic programming algorithm requires exactly 39 additions and completes essentially instantaneously, executing in microseconds. The memory usage also differs significantly: the naive approach requires a call stack depth of 40 (approximately 40 stack frames, each consuming tens of bytes), while the dynamic programming approach allocates an array of 41 integers (approximately 164 bytes for 32-bit integers or 328 bytes for 64-bit integers). The runtime behavior differs qualitatively as well. The naive algorithm's execution time grows exponentially, doubling approximately every time n increases by one, making F(50) take roughly 1,024 times longer than F(40). The dynamic programming algorithm's execution time grows linearly, so F(50) takes only about 25% longer than F(40). This comparison illustrates a fundamental lesson in algorithm design: asymptotic complexity analysis, represented by Big-O notation, accurately predicts real-world performance differences and guides algorithm selection for practical applications.

### Implementation Considerations and Trade-offs

Beyond theoretical complexity, practical implementation reveals additional considerations. The naive recursive algorithm, despite its inefficiency, requires minimal code and no explicit memory management beyond the system's call stack. However, for large n, it risks stack overflow errors when the recursion depth exceeds system limits. The dynamic programming algorithm requires explicit array allocation and index management but provides predictable memory usage and performance. Modern compilers may optimize tail recursion, though the Fibonacci recursion is not tail-recursive and cannot benefit from such optimizations. For production systems, a hybrid approach might use memoization (caching previously computed results during recursion) to maintain the recursive structure's elegance while achieving dynamic programming's efficiency. Alternatively, a space-optimized iterative solution storing only the two most recent values would achieve O(1) space complexity while maintaining O(n) time complexity. For extremely large n, matrix exponentiation methods can compute Fibonacci numbers in O(log n) time using repeated squaring, demonstrating that even linear time can be improved through mathematical insight. These alternatives illustrate how algorithm design involves balancing multiple objectives including time efficiency, space efficiency, code simplicity, and maintainability based on specific application requirements.

### Implementation: Fibonacci Algorithms with Performance Analysis

```python
import time
import sys

# Global counter to track function calls
call_count = 0

def fib_naive(n):
    """
    Pseudocode A: Naive Recursive Fibonacci
    
    This function implements the mathematical definition directly.
    Each call generates two recursive calls, creating exponential growth.
    
    Time Complexity: O(2^n) - exponential
    Space Complexity: O(n) - maximum recursion depth
    """
    global call_count
    call_count += 1  # Increment counter to track total calls
    
    # Base case: F(0) = 0, F(1) = 1
    if n <= 1:
        return n
    
    # Recursive case: F(n) = F(n-1) + F(n-2)
    # This creates redundant computation as the same values are calculated multiple times
    return fib_naive(n - 1) + fib_naive(n - 2)


def fib_dp(n):
    """
    Pseudocode B: Dynamic Programming Fibonacci
    
    This function builds the solution bottom-up, computing each value once.
    Uses an array to store intermediate results, eliminating redundancy.
    
    Time Complexity: O(n) - linear
    Space Complexity: O(n) - array storage
    """
    # Base cases
    if n <= 1:
        return n
    
    # Initialize array to store Fibonacci numbers
    # F[i] will store the i-th Fibonacci number
    F = [0] * (n + 1)  # Create array of size n+1, initialized with zeros
    F[0] = 0           # F(0) = 0
    F[1] = 1           # F(1) = 1
    
    # Build solution iteratively from bottom up
    # Each iteration computes one new Fibonacci number
    for i in range(2, n + 1):
        F[i] = F[i - 1] + F[i - 2]  # Current number is sum of previous two
    
    # Return the n-th Fibonacci number
    return F[n]


# Demonstrate both algorithms for small n
print("="*70)
print("FIBONACCI COMPUTATION: ALGORITHM DEMONSTRATION")
print("="*70)

# Test with small value to show both algorithms
test_n = 10
print(f"\nComputing F({test_n}):")

# Test naive recursion
call_count = 0  # Reset counter
start_time = time.perf_counter()
result_naive = fib_naive(test_n)
time_naive = time.perf_counter() - start_time
calls_naive = call_count

# Test dynamic programming
start_time = time.perf_counter()
result_dp = fib_dp(test_n)
time_dp = time.perf_counter() - start_time

print(f"  Naive Recursion: F({test_n}) = {result_naive}")
print(f"    - Function calls: {calls_naive}")
print(f"    - Time: {time_naive*1e6:.2f} microseconds")

print(f"  Dynamic Programming: F({test_n}) = {result_dp}")
print(f"    - Iterations: {test_n - 1}")
print(f"    - Time: {time_dp*1e6:.2f} microseconds")
print(f"  Speedup factor: {time_naive/time_dp:.1f}x")
print()

# Verify both algorithms produce the same result
assert result_naive == result_dp, "Algorithms should produce the same result!"
print("✓ Both algorithms produce the same result")
```

### Performance Analysis for n = 40

```python
print("="*70)
print("PERFORMANCE ANALYSIS FOR n = 40")
print("="*70)

target_n = 40

# For naive recursion with n=40, we'll estimate performance
# Actually running it would take too long (several minutes)
# We'll calculate theoretical number of calls

def count_fib_calls(n):
    """Calculate the exact number of function calls for naive recursion"""
    if n <= 1:
        return 1
    # Number of calls for fib(n) = calls for fib(n-1) + calls for fib(n-2) + 1
    # This follows the same pattern as Fibonacci but shifted
    calls = [1, 1]  # Base cases: fib(0) and fib(1) each require 1 call
    for i in range(2, n + 1):
        calls.append(calls[i-1] + calls[i-2] + 1)
    return calls[n]

# Calculate theoretical performance for naive recursion
theoretical_calls = count_fib_calls(target_n)
estimated_time_naive = theoretical_calls * 50e-9  # Assume ~50 nanoseconds per call

print(f"\nNaive Recursion (Theoretical Analysis for n={target_n}):")
print(f"  Total function calls: {theoretical_calls:,}")
print(f"  Estimated runtime: {estimated_time_naive:.3f} seconds ({estimated_time_naive:.2f} s)")
print(f"  Space (call stack depth): {target_n} stack frames")
print(f"  Growth rate: O(φ^n) where φ ≈ 1.618 (golden ratio)")

# Actually run dynamic programming for n=40
# This will execute almost instantaneously
start_time = time.perf_counter()
result_dp_40 = fib_dp(target_n)
time_dp_40 = time.perf_counter() - start_time

print(f"\nDynamic Programming (Actual Execution for n={target_n}):")
print(f"  Result: F({target_n}) = {result_dp_40:,}")
print(f"  Actual runtime: {time_dp_40*1e6:.2f} microseconds ({time_dp_40:.9f} seconds)")
print(f"  Operations: {target_n - 1} additions")
print(f"  Space: {target_n + 1} integers ({(target_n + 1) * 8} bytes for 64-bit integers)")

# Calculate speedup
speedup = estimated_time_naive / time_dp_40
print(f"\nPerformance Comparison:")
print(f"  Speedup factor: {speedup:,.0f}x faster")
print(f"  Time saved: {estimated_time_naive - time_dp_40:.3f} seconds")
print(f"  Efficiency gain: Dynamic Programming reduces exponential O(2^n) to linear O(n)")

# Create comparison table
comparison_data = {
    'Metric': [
        'Time Complexity',
        'Space Complexity', 
        'Total Operations/Calls',
        'Estimated Runtime',
        'Memory Usage',
        'Scalability'
    ],
    'Naive Recursion': [
        'O(2^n) - Exponential',
        'O(n) - Linear stack',
        f'{theoretical_calls:,} calls',
        f'{estimated_time_naive:.3f} seconds',
        f'{target_n} stack frames',
        'Poor - doubles per increment'
    ],
    'Dynamic Programming': [
        'O(n) - Linear',
        'O(n) - Array storage',
        f'{target_n - 1} additions',
        f'{time_dp_40*1e6:.2f} μs',
        f'{(target_n + 1) * 8} bytes',
        'Excellent - linear growth'
    ]
}

df_comparison = pd.DataFrame(comparison_data)
print("\n" + "="*70)
print("DETAILED COMPARISON TABLE")
print("="*70)
print(df_comparison.to_string(index=False))
print("="*70)
```

### Mathematical Analysis: Recursion Tree and Complexity Derivation

```python
# Visualize the recursion tree for naive Fibonacci
# This demonstrates why the naive algorithm is so inefficient

def visualize_recursion_tree():
    """
    Create a visual representation of the recursion tree for fib(5)
    This shows the redundant computations that occur
    """
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Define positions for nodes in the tree
    # We'll create a tree for fib(5) to keep it manageable
    # The tree structure is: fib(n) calls fib(n-1) and fib(n-2)
    
    # Track which values are computed multiple times
    from collections import defaultdict
    node_count = defaultdict(int)
    
    def count_nodes(n, level=0):
        """Count how many times each fib(n) is called"""
        node_count[n] += 1
        if n <= 1:
            return
        count_nodes(n-1, level+1)
        count_nodes(n-2, level+1)
    
    # Count nodes for fib(5)
    count_nodes(5)
    
    # Create text summary
    ax.text(0.5, 0.95, 'Recursion Tree for fib(5)', 
            ha='center', va='top', fontsize=16, fontweight='bold',
            transform=ax.transAxes)
    
    # Show the tree structure as text (simplified)
    tree_text = """
    Level 0:                    fib(5)
                               /      \\
    Level 1:            fib(4)          fib(3)
                       /    \\          /    \\
    Level 2:      fib(3)  fib(2)  fib(2)  fib(1)
                 /   \\    /   \\    /   \\
    Level 3: fib(2) fib(1) fib(1) fib(0) fib(1) fib(0)
            /   \\
    Level 4: fib(1) fib(0)
    
    Redundant Computations:
    - fib(3) computed 2 times
    - fib(2) computed 3 times  
    - fib(1) computed 5 times
    - fib(0) computed 3 times
    
    Total function calls: 15
    Unique values needed: 6 (fib(0) through fib(5))
    
    Efficiency loss: Computing 15 values when only 6 are unique
    """
    
    ax.text(0.05, 0.85, tree_text, ha='left', va='top', fontsize=10,
            family='monospace', transform=ax.transAxes,
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    # Create a bar chart showing redundant computations
    ax2 = plt.axes([0.6, 0.1, 0.35, 0.35])
    fib_values = sorted(node_count.keys())
    counts = [node_count[i] for i in fib_values]
    
    bars = ax2.bar(fib_values, counts, color=['red' if c > 1 else 'green' for c in counts],
                   alpha=0.7, edgecolor='black')
    ax2.set_xlabel('Fibonacci Number (n)', fontsize=10)
    ax2.set_ylabel('Times Computed', fontsize=10)
    ax2.set_title('Redundant Computations in fib(5)', fontsize=11, fontweight='bold')
    ax2.set_xticks(fib_values)
    ax2.grid(axis='y', alpha=0.3)
    
    # Add value labels on bars
    for i, (val, count) in enumerate(zip(fib_values, counts)):
        ax2.text(val, count, str(count), ha='center', va='bottom', fontweight='bold')
    
    # Add legend
    from matplotlib.patches import Patch
    legend_elements = [Patch(facecolor='green', alpha=0.7, label='Computed once'),
                       Patch(facecolor='red', alpha=0.7, label='Redundant (>1 times)')]
    ax2.legend(handles=legend_elements, loc='upper right')
    
    ax.axis('off')
    plt.show()
    
    return node_count

# Create the visualization
print("Recursion Tree Analysis:")
print("="*70)
node_count = visualize_recursion_tree()

print("\nDetailed Analysis:")
print(f"Total calls for fib(5): {sum(node_count.values())}")
print(f"Unique values computed: {len(node_count)}")
print(f"Wasted computations: {sum(node_count.values()) - len(node_count)}")
print(f"Efficiency ratio: {len(node_count) / sum(node_count.values()) * 100:.1f}%")
print("\nThis demonstrates why dynamic programming is crucial:")
print("- Naive: Recomputes same values multiple times")
print("- DP: Computes each value exactly once and stores it")
```

### Complexity Formulas and Mathematical Proof

```python
# Mathematical derivation of time complexity
print("="*70)
print("MATHEMATICAL COMPLEXITY DERIVATION")
print("="*70)

print("\n1. NAIVE RECURSION - Time Complexity")
print("-" * 70)
print("Recurrence Relation:")
print("  T(n) = T(n-1) + T(n-2) + c    where c is constant work per call")
print("  T(0) = T(1) = c               base cases")
print()
print("This recurrence has the same structure as Fibonacci numbers!")
print("Therefore, T(n) grows proportionally to F(n)")
print()
print("Using Binet's Formula for Fibonacci:")
print("  F(n) = (φ^n - ψ^n) / √5")
print("  where φ = (1 + √5)/2 ≈ 1.618 (golden ratio)")
print("        ψ = (1 - √5)/2 ≈ -0.618")
print()
print("As n → ∞, the ψ^n term becomes negligible:")
print("  T(n) ≈ c × φ^n / √5")
print()
print("Therefore: T(n) ∈ Θ(φ^n) ⊂ O(2^n)")
print()

# Calculate and display golden ratio
phi = (1 + np.sqrt(5)) / 2
psi = (1 - np.sqrt(5)) / 2
print(f"Golden ratio φ ≈ {phi:.6f}")
print(f"For comparison: 2^n vs φ^n at n=40:")
print(f"  2^40 = {2**40:,}")
print(f"  φ^40 ≈ {phi**40:,.0f}")
print()

print("2. NAIVE RECURSION - Space Complexity")
print("-" * 70)
print("Space is determined by maximum call stack depth")
print("  Maximum depth = n (following fib(n) → fib(n-1) → ... → fib(1))")
print("  Therefore: S(n) ∈ O(n)")
print()

print("3. DYNAMIC PROGRAMMING - Time Complexity")
print("-" * 70)
print("Analysis of the loop:")
print("  for i = 2 to n:")
print("      F[i] = F[i-1] + F[i-2]  # Constant time operation")
print()
print("Number of iterations: n - 1")
print("Work per iteration: O(1) - one addition, two array accesses")
print("Total work: (n - 1) × O(1) = O(n)")
print()
print("Therefore: T(n) ∈ Θ(n)")
print()

print("4. DYNAMIC PROGRAMMING - Space Complexity")
print("-" * 70)
print("Array allocation:")
print("  F = array of size (n + 1)")
print("  Each element stores one integer")
print()
print("Therefore: S(n) ∈ Θ(n)")
print()
print("Note: Space can be optimized to O(1) by keeping only last 2 values")
print()

# Create a comparison table
print("\n" + "="*70)
print("COMPLEXITY SUMMARY TABLE")
print("="*70)

complexity_table = pd.DataFrame({
    'Algorithm': ['Naive Recursion', 'Dynamic Programming'],
    'Time Complexity': ['O(φⁿ) ≈ O(1.618ⁿ) ⊂ O(2ⁿ)', 'O(n)'],
    'Space Complexity': ['O(n)', 'O(n)'],
    'Class': ['Exponential', 'Linear'],
    'Practical Limit': ['n ≈ 40', 'n ≥ 10,000'],
    'Key Characteristic': ['Redundant recursion', 'Memoized results']
})

print(complexity_table.to_string(index=False))
print("="*70)

# Demonstrate growth rates
print("\nGrowth Rate Comparison (theoretical operations):")
print("-" * 70)
print(f"{'n':>5} | {'Naive O(φⁿ)':>20} | {'DP O(n)':>15} | {'Ratio':>15}")
print("-" * 70)

for n in [10, 20, 30, 40]:
    naive_ops = phi**n
    dp_ops = n
    ratio = naive_ops / dp_ops
    print(f"{n:5} | {naive_ops:20,.0f} | {dp_ops:15} | {ratio:15,.0f}x")

print("-" * 70)
print("\nConclusion: The ratio grows exponentially, making DP dramatically")
print("more efficient as n increases.")
```
