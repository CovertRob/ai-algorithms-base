Similar to Module 9's activity, I'll discuss another closed-loop system that is present today in the med-tech industry.

- An example is a closed-loop insulin delivery system that uses a continuous glucose monitor (CGM) and an insulin pump to recommend or automate dosing adjustments for a patient with Type 1 diabetes. Similar to the Module 9 activity, this is also a closed-loop system because it senses the patient\'s current state, takes an action, observes the glucose response, and then updates future actions based on feedback. In this case, evolutionary computation would make the most sense as a way to train and tune the controller during development by searching across many possible parameter settings, safety thresholds, and response rules before deployment.

- The connection to the Markov Decision Process model from Module 9 is that both systems rely on sequential decision-making under uncertainty. So although both this example and the Module 9 model are technically closed-loop systems, the MDP perspective explains how decisions are made during operation, whereas evolutionary computation helps train, tune, and optimize that controller beforehand.

## 2. Designing and Evaluating Delivery Route Plans (Lecture 10B)

In this activity, you will work with a simple delivery-planning problem to see how a genetic algorithm can represent and evaluate candidate solutions.

Imagine a delivery company must decide which of six neighborhood stops to include in a morning route. Each stop has some value, but adding too many stops can make the route inefficient.

The six possible stops are:

- Downtown
- Harbor
- Midtown
- University
- Airport
- Suburb

Use a **6-bit chromosome**, where:

- `1` means the stop is included
- `0` means the stop is not included

Example: `101100` means the route includes Downtown, Midtown, and University.

Each stop has the following value:

- Downtown = 4
- Harbor = 3
- Midtown = 5
- University = 4
- Airport = 6
- Suburb = 2

Use the following fitness rules:

- Start with the total value of all selected stops
- If more than 4 stops are included, subtract 4
- If both Downtown and Midtown are included, add 2
- If both Airport and Suburb are included, subtract 3
- If Harbor is included without Downtown, subtract 2

### (a)
Create **four different chromosomes of your own**. For each one, decode the selected stops and compute the final fitness score using the rules above.

### (b)
After evaluating all four candidates, determine:

- which route is the strongest
- which route is the weakest
- which two routes would be more likely to be selected in a genetic algorithm

### (c)
Write a short reflection explaining what this activity shows about how candidate solutions are represented and how fitness functions can reward or penalize different choices.

## 3. Genetic Algorithm Simulation in Feature Selection

In this activity, you will simulate one generation of a genetic algorithm in a feature selection setting. You will apply selection, crossover, and mutation, then analyze what the results suggest about evolutionary optimization in practice.

Suppose you are building a predictive model and must choose which features to include.

We have 6 candidate features with respective costs:

- F1 = Age = 2
- F2 = Income = 3
- F3 = Account Balance = 2
- F4 = Credit History Length = 4
- F5 = Number of Late Payments = 1
- F6 = ZIP Code = 2

Each chromosome is a 6-bit string, where:

- `1` means the feature is included
- `0` means the feature is excluded

For example, `101100` means include F1, F3, and F4.

The objective is to **minimize fitness**.

- Lower fitness = better solution
- Fitness = Total feature cost + redundancy penalty

For redundancy penalties:

- `+2` if F1 and F3 are both selected
- `+1` if F2 and F6 are both selected

Use the following initial population:

- A = `101100`
- B = `111000`
- C = `010111`
- D = `100110`

### (a) Examine the initial population

- Rank the four chromosomes from **lowest fitness to highest fitness**.
- Identify the **best** and **worst** chromosome in the initial population.
- Briefly comment on whether the best solution uses the fewest features.

### (b) Selection

Use the following selection rule and write both parent pairs clearly.

- **Parent Pair 1** = the two chromosomes with the lowest fitness
- **Parent Pair 2** = the chromosome with the lowest fitness and the chromosome with the highest fitness

### (c) Crossover

Apply **single-point crossover after the third bit** to each parent pair. This should produce **four offspring total**.

### (d) Mutation

Apply mutation using the following rule and write the final mutated offspring clearly.

- For **Offspring 1** and **Offspring 3**, flip the **last bit**
- For **Offspring 2** and **Offspring 4**, flip the **second bit**

### (e) Evaluate each final offspring

For each final offspring:

- List the selected features
- Compute its fitness using the fitness function above
- Determine whether it is:
  - better than both parents
  - between the two parents
  - worse than both parents

### (f) Reflection on search behavior

Respond to the following:

- How did parent selection influence the offspring that were produced?
- What did crossover contribute in this example?
- What did mutation change that crossover alone could not change?
- Did the offspring appear to move the search toward lower-cost, less redundant subsets?
- What does this example suggest about the trade-off between exploration and exploitation?
- In a real feature selection problem with dozens or hundreds of candidate features, why might exact search be impractical?
- Why might an evolutionary method still be useful even if it does not guarantee the globally best subset?
- What practical trade-off does this activity illustrate, such as:
  - solution quality vs. computation time

#### (f) Reflection on Search Behavior

- Parent selection influenced the results by reusing `D`, the strongest chromosome, in both parent pairs. That increased the chance that useful low-cost feature combinations from `D` would carry forward into the offspring.
- Crossover contributed by recombining partial feature subsets from the parents. It created new combinations that were not identical to either parent, which is how the algorithm searches beyond the initial population.
- Mutation changed specific bits that crossover alone would not necessarily remove or add. For example, mutating Offspring 4 from `010110` to `000110` removed F2 and produced a lower-cost subset.

- In a real feature selection problem with dozens or hundreds of candidate features, exact search is impractical because the number of possible subsets grows exponentially.
- An evolutionary method is still useful even without guaranteeing the global best subset because it can search large spaces efficiently and often produce very strong approximate solutions in reasonable time. This is the entire point of real world optimization problems. Exact optimization is often impossible computationally.
- The practical trade-offs illustrated here are solution quality versus computation time, diversity versus convergence speed, and exactness versus scalability.
